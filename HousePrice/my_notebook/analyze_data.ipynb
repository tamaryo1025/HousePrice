```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib  # 日本語表示に対応させるため

# データの読み込み
df = pd.read_csv('../input/train.csv')

# SalePriceの分布を表示
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], kde=True)
plt.title('SalePriceの分布')
plt.xlabel('SalePrice')
plt.ylabel('頻度')
plt.show()

# 地上のリビングエリア(Sqft)とSalePriceの関係を表示
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('地上のリビングエリアとSalePriceの関係')
plt.xlabel('地上のリビングエリア (Sqft)')
plt.ylabel('SalePrice')
plt.show()
```
